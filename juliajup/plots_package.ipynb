```julia
using Plots #standard plotting library in Julia
```

```julia
x = 1:10;
y = rand(10, 1);
plot(x, y)                  
```

```julia
x = 1:0.01:10*π
y = sin.(x)

plot(x, y, label="sin(x)")
plot!(xlab="x", ylab="f(x)")    
```

```julia
x = 1:0.01:10*π
y = sin.(x)

plot(x, y, label="sin(x)",
    xlab="x", ylab="f(x)")  

```

```julia
y2=cos.(x)
plot!(x, y2, label="cos(x)", color=:red)
```
