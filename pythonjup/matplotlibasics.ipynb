```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import style
import math
style.use("dark_background")
```

```python
x=[1,2,3,4,5]
y=[16,35,40,78, 121]
z=[12, 40, 67, 74, 110]
```

```python
plt.plot(x,y)
plt.plot(x,y, 'go')
```

```python
plt.plot(x,y, 'go', label='GreenDots')
plt.plot(x,z, 'b*', label='Bluestars')
plt.title('A Simple Scatterplot')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()  # legend text comes from the plot's label parameter.
```

```python
plt.figure(figsize=(10,7)) # 10 is width, 7 is height
plt.plot(x,y, 'go', label='GreenDots')
plt.plot(x,z, 'b*', label='Bluestars')
plt.title('A Simple Scatterplot')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
```

```python
xx= np.arange(0, math.pi*2, 0.05)
yy= np.sin(xx)
plt.plot(xx,yy)
plt.xlabel("angle")
plt.ylabel("sine")
plt.title('sine wave')
```

```python
x=np.arange(1,11)
a=(x**2)
b=(x**3)
c=(x**4)
d=(x**5)
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax2 = fig.add_axes([0.72, 0.72, 0.16, 0.16])
```

```python
fig = plt.figure()

#creates a new set of axes (ax2) that shares the x-axis with ax1, 
#but can have a separate y-axis 
#similarly, twiny would return a 
#second set of axes sharing the y-axis, but with a separate x-axis. 

ax1 = fig.add_subplot(1, 1, 1)
ax2 = ax1.twinx()

t = np.linspace(0., 10., 100)
ax1.plot(t, t ** 2, 'g-')
ax2.plot(t, 1000 / (t + 1), 'r-')
ax1.set_ylabel('Density (cgs)', color='red')
ax2.set_ylabel('Temperature (K)', color='green')
ax1.set_xlabel('Time (s)')
```

```python
X = np.linspace(0, 2 * np.pi, 100)
Ya = np.sin(X)
Yb = np.cos(X)
plt.plot(X, Ya)
plt.plot(X, Yb)
plt.show()
```
